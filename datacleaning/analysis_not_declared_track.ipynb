```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/not_declared_track_survey.csv", encoding="utf-8")
df.head()
```

```python
print("Shape:", df.shape)
print("\nColumns:")
for i, col in enumerate(df.columns):
    print(f"{i}: {col}")
```

```python
df = df.rename(columns={
    "Timestamp": "timestamp",
    "What is your current year of study?": "year_of_study",
    "Have you decided on your specialization track?": "decided_track",
    df.columns[3]: "understand_roles",
    "Before choosing your track, did you feel like you had enough exposure to real-world examples?": "enough_exposure",
    "Which of the following sources do you rely on MOST to understand different tracks? (Select up to 2)": "sources_used",
    df.columns[6]: "guidance_tailored",
    df.columns[7]: "guidance_helpful",
    "What is your biggest challenge when choosing a Specialization Track?": "biggest_challenge",
    "Which of the following do you feel is MOST lacking in your decision process?": "most_lacking",
    df.columns[10]: "simulation_useful",
    df.columns[11]: "recommendation_useful"
})
df.columns.tolist()
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df["response_date"] = df["timestamp"].dt.date
```

```python
rating_cols = [
    "understand_roles",
    "guidance_tailored",
    "guidance_helpful",
    "simulation_useful",
    "recommendation_useful"
]

for col in rating_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df[rating_cols].info()
df[rating_cols].head()
```

```python
df["sources_list"] = df["sources_used"].fillna("").apply(
    lambda x: [item.strip() for item in x.split(";") if item.strip() != ""]
)
df[["sources_used", "sources_list"]].head()
```

```python
sources_exploded = df[["timestamp", "response_date", "year_of_study", "decided_track", "sources_list"]].explode("sources_list")
sources_exploded = sources_exploded.rename(columns={"sources_list": "source"})
sources_exploded = sources_exploded[sources_exploded["source"].notna() & (sources_exploded["source"] != "")]
sources_exploded.head()
```

```python
source_counts = sources_exploded["source"].value_counts().reset_index()
source_counts.columns = ["source", "count"]

source_counts
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
df["timestamp"] = df["timestamp"].dt.tz_localize(None)

sources_exploded["timestamp"] = pd.to_datetime(sources_exploded["timestamp"], errors="coerce")
sources_exploded["timestamp"] = sources_exploded["timestamp"].dt.tz_localize(None)

import os

os.makedirs("cleansed_data", exist_ok=True)

df.to_csv("cleansed_data/cleaned_track_survey.csv", index=False)
sources_exploded.to_csv("cleansed_data/sources_exploded_for_tableau.csv", index=False)
source_counts.to_csv("cleansed_data/source_counts.csv", index=False)    
                     
# import os

# os.makedirs("cleansed_data", exist_ok=True)

# df.to_excel("cleansed_data/cleaned_track_survey.xlsx", index=False)
# sources_exploded.to_excel("cleansed_data/sources_exploded_for_tableau.xlsx", index=False)
# source_counts.to_excel("cleansed_data/source_counts.xlsx", index=False)
```
